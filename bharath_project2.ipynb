
### Features to Consider Keeping

1. **`make` and `model`:** Essential for any car pricing model as different makes and models have different values.
   
2. **`year`:** Age of the car is a significant factor in pricing.

3. **`body_type`:** Different body types appeal to different buyers and can affect price.

4. **`miles`:** The mileage of a car is a key indicator of its use and condition.

5. **`engine_vol`, `engine_size`, `engine_size_unit`:** Specifications related to the car's engine can influence its value.

6. **`transmission`:** Type of transmission (manual, automatic) can affect the price.

7. **`fuel_type`:** Different fuel types can affect both the desirability and price of a car.

8. **`num_owner`:** The number of previous owners might impact a car's value.

9. **`full_service`, `part_service`, `part_warranty`, `full_dealership`:** Service history and warranty details can significantly impact the price.

10. **`first_year_road_tax`:** This might be relevant in some markets.

11. **`finance_available`:** If financing options are available for the car, it might affect its attractiveness to buyers.

### Features to Consider Dropping

1. **`Unnamed: 0`:** Seems like an index or identifier with no predictive value.

2. **`car_badges`, `car_title`, `car_sub_title`, `car_attention_grabber`:** These are likely textual descriptions that might require advanced NLP techniques to be useful.

3. **`car_specs`:** If this is a textual description, it might be complex to use without NLP.

4. **`car_seller`, `car_seller_rating`, `car_seller_location`:** Seller information may not be universally useful, unless there's a known bias (like dealers vs. private sellers).

5. **`reg`:** If this refers to the registration plate, it's usually not useful for pricing.

6. **`ulez`:** Ultra-Low Emission Zone compliance is very specific and may not be widely relevant.

7. **`brand_new`:** Since it's a used car dataset, this might not be relevant unless it signifies very recent models.

8. **`discounted`:** Discount information might be transient and not indicative of the car's actual value.

# Data Modelling

```python
#creating the new data with only required columns
data4 = data3[['make', 'model', 'miles', 'age', 'body_type', 'transmission', 'engine_vol',
               'feul_type', 'num_owner', 'car_price']]
data4.head()
```

```python
import pandas as pd
import numpy as np
import time
from tqdm import tqdm
import matplotlib
import matplotlib.pyplot as plt

#libraries for preprocessing
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#libraries for evaluation
from sklearn.metrics import mean_squared_log_error,r2_score,mean_squared_error
from sklearn.model_selection import train_test_split


#libraries for models
from sklearn.linear_model import LinearRegression

from sklearn.linear_model import Ridge
from sklearn.linear_model import LassoCV,RidgeCV
from sklearn.model_selection import GridSearchCV

from yellowbrick.regressor import AlphaSelection

from sklearn.utils.metaestimators import available_if
from sklearn.metrics import DistanceMetric
from sklearn.linear_model import Lasso

from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

from sklearn.ensemble import AdaBoostRegressor

import xgboost as xgb

import lightgbm as lgb 


import warnings
warnings.filterwarnings('ignore')
from sklearn import set_config
set_config(display="diagram")
```

```python
#function to split dataset int training and test
def trainingData(data4,n):
    X = data4.iloc[:,n]
    y = data4.iloc[:,-1:].values.T
    y=y[0]
    X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.9,test_size=0.1,random_state=0)
    return (X_train,X_test,y_train,y_test)

X_train,X_test,y_train,y_test=trainingData(data4,list(range(len(list(data4.columns))-1)))
```

```python
X_train
```

```python
#some of models will predict neg values so this function will remove that values
def remove_neg(y_test,y_pred):
    ind=[index for index in range(len(y_pred)) if(y_pred[index]>0)]
    y_pred=y_pred[ind]
    y_test=y_test[ind]
    y_pred[y_pred<0]
    return (y_test,y_pred)

#function for evaluation of model
def result(y_test,y_pred):
    r=[]
    r.append(mean_squared_log_error(y_test, y_pred))
    r.append(np.sqrt(r[0]))
    r.append(r2_score(y_test,y_pred))
    r.append(round(r2_score(y_test,y_pred)*100,4))
    return (r)

#dataframe that store the performance of each model
accu=pd.DataFrame(index=['MSLE', 'Root MSLE', 'R2 Score','Accuracy(%)'])    
```

# 1)Linear Regression

```python
#fitting model
LR=LinearRegression()
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
```

```python
#calculating error/accuracy
y_test_1,y_pred_1=remove_neg(y_test,y_pred)
r1_lr=result(y_test_1,y_pred_1)
print('Coefficients: \n', LR.coef_)
print("MSLE : {}".format(r1_lr[0]))
print("Root MSLE : {}".format(r1_lr[1]))
print("R2 Score : {} or {}%".format(r1_lr[2],r1_lr[3]))
accu['Linear Regression']=r1_lr
```

```python
#Ploting feature importance graph
coef = pd.Series(LR.coef_, index = X_train.columns)
imp_coef = coef.sort_values()
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Linear Regression Model")
plt.savefig('D:/Bharath Project/Images/Linear-Regression-Feature-Importance.jpg')
plt.show()
```

# 4)KNN

```python
#estimating MSLE for k=1-9
R_MSLE=[]
for i in range(1,10):
    KNN=KNeighborsRegressor(n_neighbors=i)
    KNN.fit(X_train,y_train)
    y_pred=KNN.predict(X_test)
    error=np.sqrt(mean_squared_log_error(y_test, y_pred))
    R_MSLE.append(error)
    print("K =",i," , Root MSLE =",error)
```

```python
#plotting error

curve = pd.DataFrame(R_MSLE) #elbow curve 
plt.figure(figsize=(8,4))
plt.xticks(list(range(1,10)), list(range(1,10)), rotation='horizontal')
plt.plot(list(range(1,10)),R_MSLE)
plt.xlabel('K')
plt.ylabel('MSLE')
plt.title('Error Plot for Each K')
plt.savefig('D:/Bharath Project/Images/KNN-Error-Plot.jpg')
plt.show()
```
